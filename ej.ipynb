Ecuación del péndulo

Ejemplo

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parámetros del péndulo
g = 9.81  # gravedad (m/s^2)
l = 1.0   # longitud (m)

# Ecuación diferencial del péndulo
def pendulo(t, y):
    theta, omega = y
    dtheta_dt = omega
    domega_dt = - (g / l) * np.sin(theta)
    return [dtheta_dt, domega_dt]

# Condiciones iniciales: [ángulo inicial, velocidad angular inicial]
y0 = [np.pi / 4, 0.0]

# Intervalo de tiempo
t_span = (0, 10)
t_eval = np.linspace(*t_span, 250)

# Resolver la ecuación diferencial
sol = solve_ivp(pendulo, t_span, y0, t_eval=t_eval)

# Graficar el ángulo en función del tiempo
plt.plot(sol.t, sol.y[0])
plt.xlabel('Tiempo (s)')
plt.ylabel('Ángulo (rad)')
plt.title('Movimiento de un péndulo simple')
plt.grid()
plt.show()
```

La gráfica anterior muestra cómo varía el ángulo del péndulo simple en función del tiempo. El eje horizontal representa el tiempo en segundos, mientras que el eje vertical indica el ángulo (en radianes) respecto a la posición de equilibrio. Se observa un movimiento oscilatorio, característico de un péndulo, donde el ángulo aumenta y disminuye periódicamente debido a la acción de la gravedad. La amplitud y la frecuencia de las oscilaciones dependen de las condiciones iniciales y de los parámetros físicos del sistema, como la longitud del péndulo y la aceleración gravitacional.
